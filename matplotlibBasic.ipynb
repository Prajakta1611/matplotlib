```python
# Write a Python program to draw a line with suitable label in the x axis, y axis and a title.
```

```python
import matplotlib.pyplot as plt
```

```python
x = [2,4,6,8,10]
y = [1,2,3,4,5]
```

```python
plt.plot(x,y)
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.title("Line")
plt.show()
```

```python
# Write a Python program to plot two or more lines with legends, different widths and colors.
```

```python
x1 = [10,20,30]
y1 = [20,40,10]
```

```python
x2 = [10,20,30]
y2 = [40,10,30]
```

```python
plt.plot(x1,y1,c="c" ,linewidth="3",label="x1,y1")
plt.plot(x2,y2,c="g",linewidth="2",label="x12,y2")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.legend()
plt.show()
```

```python
#  Write a Python program to plot two or more lines with different styles.
```

```python
plt.plot(x1,y1,c="c" ,linewidth="3", linestyle="--" ,label="x1,y1")
plt.plot(x2,y2,c="g",linewidth="2", linestyle=":", label="x12,y2")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.legend()
plt.show()
```
